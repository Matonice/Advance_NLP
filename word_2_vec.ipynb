**Training our own model**

```python
#Read in the data and clean up the column names
import gensim
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
pd.set_option('display.max_colwidth', 100)

data = pd.read_csv("/content/SMSSpamCollection.tsv",sep='\t', header=None)
data.columns = ['label', 'text']
data.head()
```

```python
#clean data using the built in gensim function
data['text_clean'] = data['text'].apply(lambda x: gensim.utils.simple_preprocess(x))
data.head()
```

```python
"#split your data into training and testing set
x_train, x_test, y_train, y_test = train_test_split(data['text_clean'],
                                                    data['label'], test_size = 0.2)
```

```python
#Train the word2vec model
w2v_model = gensim.models.Word2Vec(x_train,
                                   size=100,
                                   window=5,
                                   min_count=2)
```

```python
#Explore the word vector for king base on our trained model
w2v_model.wv['wonderful']
```

```python
#find the most similar words to king based on words vectors from our trained model
w2v_model.wv.most_similar('king')
```

```python
w2v_model.wv.index2word
```

```python
#Generate aggregrated sentence vectors based on the word vectors for each word in the sentence
w2v_vect = np.array([np.array([w2v_model.wv[i] for i in ls if i in w2v_model.wv.index2word])
for ls in x_test])
```

```python
#why is the length of the sentence different than the lenght of the sentence vector
for index, value in enumerate(w2v_vect):
  print(len(x_test.iloc[index]), len(value))
```

```python
#Compute sentence vectors by averaging the word vectors for the words contained in the sentences
w2v_vect_avg = []

for vect in w2v_vect:
  if len(vect) !=0:
    w2v_vect_avg.append(vect.mean(axis=0))
  else:
    w2v_vect_avg.append(np.zeros(100))
```

```python
#Are our sentence vector lenghts now consistent?
for index, value in enumerate(w2v_vect_avg):
  print(len(x_test.iloc[index]), len(value))
```
